```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
```

```python
url = "https://img.freepik.com/free-photo/a-red-mini-coupe-parked-in-the-port-xenon-lights-back-view_114579-5196.jpg?t=st=1696261021~exp=1696261621~hmac=049839b03f6c350ff97238614290f57734e4f37dd70cff4f5c5d91908d0235a0"
image = io.imread(url)
plt.imshow(image)
```

```python
image_flipped = cv2.flip(image, 1)
plt.imshow(image_flipped)
```

```python
image_gauss = cv2.GaussianBlur(image_flipped, ksize=(3,3), sigmaX =0, sigmaY=0)
plt.imshow(image_flipped)
```

```python
sobelxy = cv2.Sobel(image_gauss, ddepth=-1, dx=1, dy=1, ksize=5)
plt.imshow(sobelxy)
```
